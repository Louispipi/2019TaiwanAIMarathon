```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')
```

```python
data_path='C:/Users/user/Documents/Python Scripts/AIMarathon/'
x_train=pd.read_csv(data_path+'HW48_train.csv',header=None)
y_train=pd.read_csv(data_path+'HW48_trainLabels.csv',header=None)
print ('The shape of x_train:',x_train.shape)
print ('The shape of y_train:',y_train.shape)
x_train.head(10)
```

```python
#檢查資料是否有空缺值
def missing_count(df):
    data_na=df.isnull().sum()
    data_na=data_na.drop(data_na[data_na==0].index)
    missing_count=pd.DataFrame({'Missing Count':data_na})
    display(missing_count)
missing_count(x_train)
```

```python
#把數值標準化
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=pd.DataFrame(scaler.transform(x_train))
x_train.head(10)
```

```python
#先嘗試羅吉斯回歸
from sklearn.linear_model import LogisticRegression
x_test=pd.read_csv(data_path+'HW48_test .csv',header=None)

reg=LogisticRegression()
reg.fit(x_train,y_train)
y_predict=reg.predict(x_test)
y_predict=pd.DataFrame(y_predict)
print (y_predict.shape)
y_predict.head(5)
```

```python
#輸出結果
y_predict.to_csv('y_test.csv')
```
